```python

import numpy as np
```

```python
RANDOM_SEED = 42
dataset = '../data/gestures_test_seq.csv'
model_save_path = '../data/gesture_sequence_classifier_test.hdf5'
```

```python
ds = []
with open(dataset) as f:
    lines = f.readlines()
    for line in lines:
        spl = line.split(" | ")
        gesture = spl[0]
        coords = list(map(eval, spl[1:]))
        ds.append((gesture, coords))
```

```python
for pair in ds:
    coords_seq = pair[1]
    for coords in coords_seq:
        first_x = coords[0]
        first_y = coords[1]
        coords = np.array(coords)
        coords[0::2] -= first_x
        coords[1::2] -= first_y
ds = list(map(lambda x: (x[0], np.array(x[1])), ds))
```
